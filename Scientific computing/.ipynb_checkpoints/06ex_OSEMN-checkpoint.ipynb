1\. **Text files**

Perform the following operations on plain `txt` files:

+ create a list of integrer numbers and then save it to a text file named `data_int.txt`. Run the `cat` command to print the content of the file.
+ create a matrix of 5x5 floats and then save it to a text file named `data_float.txt`. Use the `cat` command to print the content of the file.
+ load the `txt` file of the previous point and convert it to a `csv` file by hand.

````python
l = [x for x in range(10)]
out_file_name = "data_int.txt"

with open(out_file_name, 'w') as outfile:
    for x in l:
        outfile.write(str(x)) # write last line to a file
    outfile.write("\n")

!cat "data_int.txt"
````

````python
import numpy.random as npr
npr.seed(123)

matrix = npr.randn(5, 5)
out_file_name = "data_float.txt"

with open(out_file_name, 'w') as outfile:
    for row in matrix:
        for x in row:
            #print(x)
            outfile.write(str(x)+" ") # write last line to a file
        outfile.write("\n") 
        
!cat "data_float.txt"
````

````python
file_name = "data_float.txt"

list_word = []
with open(file_name, mode = 'r') as file:
    for line in file:
        for word in line.split():
            list_word.append(word)
print(list_word)          
````

````python

with open('file.csv', mode = 'w') as file:
    for word in list_word[:-1]:
        file.write(word+',')
    file.write(list_word[-1])
  
!cat "file.csv"
````
