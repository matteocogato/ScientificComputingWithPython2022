1\. **Spotting correlations**

Load the remote file:

```bash
https://www.dropbox.com/s/aamg1apjhclecka/regression_generated.csv
```

with Pandas and create scatter plots with all possible combinations of the following features:
    
  + features_1
  + features_2
  + features_3
  
Are these features correlated?

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
#%matplotlib inline

file_name= "regression_generated.csv"
data = pd.read_csv(file_name)

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(12, 5))

ax1.scatter(x=data['features_1'], y=data['features_2'], marker='o', c='r')
ax1.set_title('Features_1 vs features_2', fontsize=14,  fontweight ="bold", pad=10)
ax1.set_xlabel('features_1')
ax1.set_ylabel('features_2')

ax2.scatter(x=data['features_1'], y=data['features_3'], marker='o', c='b')
ax2.set_title('Features_1 vs features_3', fontsize=14, fontweight ="bold", pad=10)
ax2.set_xlabel('features_1')
ax2.set_ylabel('features_3')

ax3.scatter(x=data['features_2'], y=data['features_3'], marker='o', c='g')
ax3.set_title('Features_2 vs features_3', fontsize=14,  fontweight ="bold",pad=10)
ax3.set_xlabel('features_2')
ax3.set_ylabel('features_3')

fig.tight_layout(w_pad=4)

#these features are not corrlated
````

````python
df = pd.DataFrame(data, columns=['features_1', 'features_2','features_3'])
sns.pairplot(df)
data
df
````
